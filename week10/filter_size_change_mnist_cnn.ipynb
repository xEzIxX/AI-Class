**실험1: 기존 mnist_cnn.ipynb 파일에서 필터 크기를 변경**

**[변경 사항]**
- 첫 번째 Conv2D의 output shape는 (26, 26, 32)에서 (24, 24, 32)로 감소  
- Parameter 수는 첫 번째 Conv2D layer에서 320개에서 832개로 증가  
- 전체 parameter 수도 93,322개에서 93,834개로 증가

```python
# 전처리
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

# MNIST 데이터 불러오기
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# 픽셀 값을 0~1 사이로 정규화 한다.
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
# 컨벌루션 신경망 생성

model = models.Sequential()

# MNIST 숫자 이미지 모양인 28, 28, 1 형상의 입력을 처리하는 신경망을 정의
# 실험1: 첫 번째 필터의 크기를 3*3 -> 5*5로 변경
model.add(layers.Conv2D(32, (5, 5), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))

model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
# 모델 출력
model.summary()
```

```python
# 컴파일과 훈련
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(train_images, train_labels, epochs=5);
```
