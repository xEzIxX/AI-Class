**실험3: 기존 mnist_cnn.ipynb 파일에서 Dense layer를 변경**

**[변경 사항]**
- Conv2D layer 수가 3개에서 4개로 증가
- 모델의 마지막 Conv2D output shape가 (3, 3, 64)에서 (1, 1, 128)로 변경
- Flatten layer의 output shape가 (576)에서 (128)로 감소
- 추가된 Conv2D layer의 Parameter 수는 73,856개
- Dense layer의 Parameter 수는 36,928개에서 8,256개로 감소
- 전체 Parameter 수는 93,322개에서 138,506개로 증가

```python
# 전처리
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

# MNIST 데이터 불러오기
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# 픽셀 값을 0~1 사이로 정규화 한다.
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
# 컨벌루션 신경망 생성

model = models.Sequential()

# MNIST 숫자 이미지 모양인 28, 28, 1 형상의 입력을 처리하는 신경망을 정의
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))

model.add(layers.Flatten())
# 실험3: Dense layer 노드 수 64 -> 128로 변경
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
# 모델 출력
model.summary()
```

```python
# 컴파일과 훈련
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(train_images, train_labels, epochs=5);
```
