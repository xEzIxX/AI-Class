```python
import pandas as pd
import numpy as np;

file_path = '/content/drive/MyDrive/Colab Notebooks/wine.csv';

# 데이터 읽기
df = pd.read_csv(file_path);

# 상위 5개 행 확인
# print(df.head());
df.head()
```

```python
# 정답(label) 데이터 설정
# Alcohol = 알콜
y=df['Alcohol'];
print(y.value_counts());
```

```python
# 입력 데이터(feature) 생성
# 알콜을 예측해야하니 해당 값을 제외
X=df.drop('Alcohol',axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split

# 학습 데이터 / 테스트 데이터 분리
# 25퍼센트를 테스트 케이스로 준다
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
# 평균 제곱근 오차를 구하는 메서드 정의
def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

# 성능 평가
print("\n--- Linear Regressor ---")
print('평균제곱근오차', mean_squared_error(ly_preds, y_test))
# print('평균제곱근오차', mse_np(ly_preds, y_test))
# print('평균제곱근오차', mse(ly_preds, y_test))
```

```python
# Decision Tree 회귀 모델
# feature 기준으로 데이터를 반복 분할하여
# 값의 평균을 이용해 연속값(BMI)을 예측하는 트리 기반 회귀 모델

from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=0)

# 모델 학습
dt_model.fit(X_train, y_train)

# 예측
dt_preds = dt_model.predict(X_test)

# 성능 평가 (MSE)
print("\n--- Decision Tree Regressor ---")
print("평균제곱오차", mean_squared_error(y_test, dt_preds))
```

```python
# Random Forest 회귀 모델
# 여러 개의 Decision Tree를 생성하고
# 각 트리의 예측값 평균을 이용해 연속값을 예측하는 모델

from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(random_state=0)

# 모델 학습
rf_model.fit(X_train, y_train)

# 예측
rf_preds = rf_model.predict(X_test)

# 성능 평가 (MSE)
print("\n--- Random Forest Regressor ---")
print("평균제곱오차", mean_squared_error(y_test, rf_preds))
```
