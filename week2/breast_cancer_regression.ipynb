```python
import pandas as pd
import numpy as np;

# 파일 경로 설정
# 이미 전문가에 의해 특징이 추출된 데이터
file_path = '/content/drive/MyDrive/Colab Notebooks/breast_cancer.csv';

# 데이터 읽기
df = pd.read_csv(file_path);

# 상위 5개 행 확인
# print(df.head());
df.head()
```

```python
y=df['worst compactness'];
print(y.value_counts());
```

```python
y
```

```python
df2=df.drop('worst compactness',axis=1) # worst compactness 열 삭제
df2.head()
```

```python
from sklearn.model_selection import train_test_split

# 25퍼센트를 테스트 케이스로 준다
# 75 퍼센트를 학습, 정답값이 없는 25퍼센트로 테스트 진행
X_train, X_test, y_train, y_test = train_test_split(df2, y,test_size=0.25,random_state=0)
```

```python
X_test
```

```python
# worst compactness 값의 오차를 추론

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train) # 정답값 (worst compactness) 이 있는 train으로 학습
ly_preds = model.predict(X_test) # 정답값이 없는 X_test로 테스트 진행

# 평균 제곱근 오차를 세번 구하는 이유: 다양한 연산 방법을 사용해도 동일하게 나옴을 보여주기 위해

print("\n[Linear Regression Regressor 의 결과]\n")

# 방식 1: 사이킷런 표준 함수 사용
print('평균제곱근오차', mean_squared_error(ly_preds, y_test))

# 방식 2: 넘파이 배열 연산 활용
# (실제값 - 예측값)^2 의 평균을 구하는 MSE 수식 적용
def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y_test))

# 방법3: 파이썬 반복문-> 하나씩 꺼내서 오차를 더해감
def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor

# 1. 모델 생성 및 학습
dt_model = DecisionTreeRegressor(random_state=0)
dt_model.fit(X_train, y_train) # 75% 학습 데이터로 나무 규칙 생성

# 2. 예측 진행
dt_preds = dt_model.predict(X_test) # 25% 테스트 데이터로 수치 추론

print("\n[Random Forest Regressor 의 결과]\n")

# 방식 1: 사이킷런 표준 함수 사용
print('평균제곱근오차', mean_squared_error(dt_preds, y_test))

# 방식 2: 넘파이 배열 연산 활용
print('평균제곱근오차)', mse_np(dt_preds, y_test))

# 방식 3: 파이썬 반복문 활용
print('평균제곱근오차', mse(dt_preds, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

# 1. 모델 생성 및 학습
rf_model = RandomForestRegressor(random_state=0)
rf_model.fit(X_train, y_train) # 여러 개의 나무가 각자 학습 진행

# 2. 예측 진행
rf_preds = rf_model.predict(X_test) # 모든 나무의 예측값 평균을 최종 답으로 채택

print("\n[Random Forest Regressor 의 결과]\n")

# 방식 1: 사이킷런 표준 함수 사용
print('평균제곱근오차', mean_squared_error(rf_preds, y_test))

# 방식 2: 넘파이 배열 연산 활용
print('평균제곱근오차', mse_np(rf_preds, y_test))

# 방식 3: 파이썬 반복문 활용
print('평균제곱근오차', mse(rf_preds, y_test))
```
